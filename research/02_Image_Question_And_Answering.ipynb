```python
%%time
image = load_image_from_url("https://m.media-amazon.com/images/W/MEDIAX_792452-T2/images/I/71s3aJbIRkL._AC_UF1000,1000_QL80_.jpg", (500, 500))
# Prepare the contents
prompt_1 = "What is shown in the picture?"
prompt_2 = "What these characters are doing?"
contents = [prompt_1, prompt_2, image]

responses = vision_model.generate_content(contents, stream=True)

print("\n-------Response--------")
for response in responses:
    print(response.text, end="")
```
